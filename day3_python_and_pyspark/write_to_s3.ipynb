```python
#Mount S3
dbutils.fs.mount(SOURCE_URL, MountName)
```

```python
# display sample data to read
display(dbutils.fs.ls('/databricks-datasets'))
```

```python
#read the csv 
df = spark.read.csv("dbfs:/databricks-datasets/adult/", header=False)
# rename columns instead of using 
df = df.withColumnRenamed("_c0","Age")\
       .withColumnRenamed("_c1","sector")\
       .withColumnRenamed("_c2","Phone Number")\
       .withColumnRenamed("_c3","Education Level")\
       .withColumnRenamed("_c4","Education Number")\
       .withColumnRenamed("_c5","Marital Status")\
       .withColumnRenamed("_c6","Occupation")\
       .withColumnRenamed("_c7","Relation")\
       .withColumnRenamed("_c8","Race")\
       .withColumnRenamed("_c9","Gender")\
       .withColumnRenamed("_c10","Capital Gain")\
       .withColumnRenamed("_c11","Capital Loss")\
       .withColumnRenamed("_c12","Hours per week")\
       .withColumnRenamed("_c13","Native Country")\
       .withColumnRenamed("_c14","Income")\
     
```

```python
# limit data to write to s3
df = df.limit(5)
```

```python
# write to S3
df.write.save(f'/mnt/buckettweeq2/example2.csv',format="csv")
```

```python
#read the file on S3 bucket

file_location = '/mnt/buckettweeq2/example2.csv'
file_type = "csv"
# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = "",""
# The applied options are for CSV files. For other file types, these will be ignored.
df1 = spark.read.format(file_type) \
.option("inferSchema", infer_schema) \
.option("header", first_row_is_header) \
.option("sep", delimiter) \
.load(file_location)
display(df1)
```
